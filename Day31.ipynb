```python
#array oluşturma***
arr2=np.array([[1,2,3],[4,5,6]])
print(arr2)
type(arr2) 
```

```python
#arange nedir?
arr3=np.arange(0,11)  #0 dan 10 a kadar sayılar oluşturur 11 dahil değil
print(arr3)
type(arr3)
#belli aralıklarla da oluşturabiliriz
arr4=np.arange(0,21,2)  #0 dan 20 ye kadar
print(arr4)
#2 şer 2 şer artar
type(arr4)
```

```python
#linspace nedir? belli bir aralıkta eşit aralıklı sayılar oluşturur
arr5=np.linspace(0,10,5)  #0 ile 10 arasında 5 sayı oluşturur
print(arr5)
type(arr5)

arr6=np.linspace(0,10,20)  #0 ile 10 arasında 20 sayı oluşturur
print(arr6)
type(arr6)
```
