```python
my_dictionary={"James": [40,30,np.nan], "Alice": [60, np.nan, 70], "Bob": [70,80,90],"Merry":[80,90,100]}
```

```python
my_data_frame=pd.DataFrame(my_dictionary)#sözlüğümüzü aldık dataframe yaptık yani tablo haline getirdik
```

```python
my_data_frame
```

```python
my_data_frame.dropna()#içinde nan yani boş olan satırları atar
#nan sütunları silmek istersen axis=1 parametresini verirsin içine
```

```python
my_data_frame.dropna(axis=1,thresh=3)#kaldır diyoruz
#neyi kaldır axis 1 verdik sütunu kaldıracak thresh 3 demek en az 3 tane veri olan sütunları tut gerisini at anlamına gelir
```

```python
#groupby görelim gruplandırma yapacağız
```

```python
salary_dict={"Programming Language":["Python","Java","Python","C++","Java","C++","Python"],"Name":["Ali","Veli","Ayşe","Fatma","Can","Deniz","Ece"],"Salary":[6000,5500,7000,6500,5200,7200,7100]}
```

```python
salary_frame=pd.DataFrame(salary_dict)
```

```python
salary_frame
```

```python
group_object=salary_frame.groupby("Programming Language")
#programlama diline göre gruplandır
```

```python
#group_object. yap . dan sonra fonskiyonlar gelir
group_object.count()
```

```python
group_object.describe()#özet istatistik bilgileri verir
```
