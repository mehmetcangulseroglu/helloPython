```python
#görselleştirme kütüphanesidir.
#numpy pandas vs ile birlikte kullanılır.
import matplotlib.pyplot as plt#veya import matplotlib as mpl dersin ama pyplot yeterli çizim için
import numpy as np
```

```python
age_list=[25,50,34,12,53,65,23,13,40]
weight_list=[50,80,60,70,90,100,45,55,70]
```

```python
#normal liste de olabilir numpy array de olabilir
age_list_numpy=np.array(age_list)
weight_list_numpy=np.array(weight_list)
```

```python
#plot çizim işlemini başlatır içine x ve y değerlerini veririz
#eğer istersen plt.plot ile grafiği çizerken parametre olarak çubuk renkleri, çizgi kalınlığı vs verebiliriz
plt.plot(age_list_numpy,weight_list_numpy,"r")
#x ve y eksenlerine isim verebiliriz
plt.xlabel("age")
plt.ylabel("weight")
#grafiğe başlık ekleyebiliriz
plt.title("age vs weight graph")
```

```python
my_numpy=np.array(range(0,15))
```

```python
my_numpy2=my_numpy ** 3
```

```python
my_numpy2
```

```python
plt.plot(my_numpy,my_numpy2,"g+")#renk verirken yanına o simgeyi grafiğe basar
```

```python
plt.plot(my_numpy,my_numpy2,"g*-")
```

```python
plt.plot(my_numpy,my_numpy2,"g--")
```

```python
#subplot ile birden fazla grafiği tek bir figürde gösterebiliriz
plt.subplot(1,2,1)#1 satır 2 sütunluk birinci grafik
plt.plot(my_numpy,my_numpy2,"r--")
plt.subplot(1,2,2)#1 satır 2 sütunluk ikinci grafik
plt.plot(my_numpy2,my_numpy,"g*")
```

```python
#grafik içinde grafik gibi durumlara bakalım
```

```python
#şimdiye kadar değişkene  atamadık bir yerde durmuyor hep baştan yazamayız
#figure yapılarına bakalım
my_figure=plt.figure()#boş bir figür oluşturduk
my_axes=my_figure.add_axes([0.1, 0.1, 0.5, 0.5])#figüre bir eksen ekledik
#burda son iki parametre grafiğin boyutunu belirler bundan önce biz beliryemiyorduk mesela

my_axes.plot(age_list_numpy,weight_list_numpy,"r--")
#yani sen bir figür oluşturdun ona boyut verdin ve değişkene atadın
#artık onun içine çizdirdin grafiği

#artık diğer herşeyi yapabilirim
my_axes.set_xlabel("age")
my_axes.set_ylabel("weight")
my_axes.set_title("age vs weight graph")
```

```python
#plot yaparken yani çizdirirken renk verdik ama 
#bunu color ve alpha(saydamlık) parametreleri ile de yapabiliriz
plt.plot(age_list_numpy,weight_list_numpy,color="green",alpha=0.9)
```

```python
plt.scatter(age_list_numpy,weight_list_numpy,color="blue")
#scatter ile nokta grafiği çizdiririz
```
